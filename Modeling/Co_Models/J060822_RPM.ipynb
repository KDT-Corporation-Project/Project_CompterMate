### 2. 모델 학습

```python
data = pd.read_csv('../DATA/raw_2023051820231018_경대기업맞춤형.csv')
train_data, test_data = sp().preps(data)
X_train, y_train = train_data.drop('scale_pv', axis=1), train_data['scale_pv']
X_test, y_test = test_data.drop('scale_pv', axis=1), test_data['scale_pv']

# 모델 선언 : GradientBoostingRegressor, RandomForestRegressor, LGBMRegressor, HuberRegressor, BayesianRidge / StandardScaler
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from lightgbm import LGBMRegressor
from sklearn.linear_model import HuberRegressor, BayesianRidge
from sklearn.preprocessing import StandardScaler

models = [
    GradientBoostingRegressor(),
    RandomForestRegressor(),
    LGBMRegressor(),
    HuberRegressor(),
    BayesianRidge()
]

scaler = StandardScaler()

for model in models:
    model.fit(X_train, y_train)
    scaler.fit(X_train)
    model_gb = RPMCalculator(model, scaler)
    rpm_df = model_gb.set_dataframe(test_data).reset_index(drop=True)
    rpm_df = model_gb.calculate_rpm(rpm_df)
    print(rpm_df)
    break
```
